```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../ESRGAN/LR/datasets/Data_Entry_2017_v2020.csv")
df.head()
```

```python
new_df = df[df["Finding Labels"].isin(["No Finding" , "Atelectasis"])]

mapping = {
    "No Finding" : 0,
    "Atelectasis" : 1
}

new_df = new_df[new_df["Finding Labels"].isin(mapping.keys())]
new_df["Finding Labels"] = new_df["Finding Labels"].map(mapping)
df_slice = (new_df.groupby("Finding Labels")).apply(lambda x: x.sample(n = 4000 , random_state = 42) if len(x) >= 4000 else x).reset_index(drop = True)
df_slice.head()
df_slice.to_csv("df_onehot.csv" , index = False)
```
